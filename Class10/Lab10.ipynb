## Exercise 1:

Let's have a go with the BBC website. Load the BBC News page and print out all of the links, then write a loop that visits and prints out the news titles in the page of each link.

```python

from bs4 import BeautifulSoup
import time
import requests
response = requests.get("https://abc.com")
data = response.text

data_bs = BeautifulSoup(data)

list_of_links = data_bs.find_all('a')

for l in list_of_links:
    print(l.get('href'))
    response = requests.get("https://abc.com"+ l.get('href'))
    data = response.text
    data_bs = BeautifulSoup(data)
    list_of_links_in_subpage = data_bs.find_all('a')
    for inner_l in list_of_links_in_subpage:
        print(inner_l.text)
    time.sleep(3)

```
