**Question 2 [15 marks]**

Write a python function that receives a string variable `st`, checks that it is a string type, and then creates a dictionary of the lowercase vowels, where each key is a vowel, and each value is the count of that vowel. If `st` is not a string, return an empty dictionary, `{}`.

```
def counter(st):
  if type(st) == str:
    a_count=st.count('a')
    e_count=st.count('e')
    i_count=st.count('i')
    o_count=st.count('o')
    u_count=st.count('u')

    d={"a":a_count,"e":e_count,"i":i_count,"o":o_count,"u":u_count}
    return d
  else:
    return {}

print(counter('important'))
```

**Question 3: Text encoder [25 marks]**

Write a python function encoder that receives a string text, and the does the following:

For each character in the string:
-	convert it into its numerical value
-	convert the numerical value into a string
-	for each character in the numerical value (now it is a string of the form “###”, where “#” represent a number of 0 to 9): convert that character into a corresponding string according to the following rule:
  -	0 becomes a
  -	1 becomes b
  -	2 becomes c
  -	3 becomes d
  -	4 becomes e
  -	5 becomes f
  -	6 becomes g
  -	7 becomes h
  -	8 becomes i
  -	9 becomes j

Then print out the result.


```
def num_digits(n):
  n = str(n)
  return len(n)

def padding(n):
  if num_digits(n)==2:
    return "0"+str(n)
  elif num_digits(n)==1:
    return "00"+str(n)
  else:
    return str(n)

def converter(c):
  c = int(c)
  if c == 0: 
    return 'a'
  elif c == 1: 
    return 'b'
  elif c == 2: 
    return 'c'
  elif c == 3: 
    return 'd'
  elif c == 4: 
    return 'e'
  elif c == 5: 
    return 'f'
  elif c == 6: 
    return 'g'
  elif c == 7: 
    return 'h'
  elif c == 8: 
    return 'i'
  elif c == 9: 
    return 'j'

def encoder(p):
  cipher_text=""
  for l in p:
    v=ord(l)
    v=padding(v)

    v=converter(v[0])+converter(v[1])+converter(v[2])

    cipher_text+=v
  return cipher_text

encoder("Python")

```

**Question 4: Flattening a collection [25 marks]**
 
Write a python function that receives a dictionary, and then generates a list the_list of all the keys and values of the dictionary. If the dictionary key or value is a list, your code should add the elements of that list into the_list. Your code should not use any functions except keys() and values().

```
a={4:7,(3,4):"a"}



def flattening(d):
  #create the empty list that will be populated with elements
  the_list=[]
  #iterate through every key-value pair of the dictionary
  for k in d:
    #check the type of k
    if type(k) == tuple:
      #if it is a tuple, go through each element and add it to the_list
      for i in k:
        the_list+=[i]
    #otherwise, add it to the the_list
    else:
      the_list+=[k]
    #check the type of the value
    v=d[k]
    #if it is a list, go through each element and add it to the_list
    if type(v) == list:
      for i in v:
        the_list+=[i]
    else:
    #otherwise, add it to the_list
      the_list+=[v]

  return the_list

print(flattening(a))   
    
    
    

[4,7,3,4,"a"]
```
