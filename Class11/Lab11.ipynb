## Exercise 1: 

One way to understand cryptography is to implement some parts of it ourselves. Let's write code that encodes a text string.

The `coolEncoder()` function we will write here receives a text string and then applies the following rule to it:
- Convert each character to a number.
  - If it is a lowercase letter, replace each letter by its order in the alphabet.
  - If it is an uppercase letter, add 1 to its order number
  - If it is not a letter, just put 0 before the character.

For example, `"Abc 134"` will be expressed as `"0202030 010304"`.

```
def num_digits(n):
  if n < 10:
    return 1
  else:
    return 2

def padding(n):
  if num_digits(n)==1:
    return "0"+str(n)
  else:
    return str(n)

caps="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
smalls="abcdefghijklmnopqrstuvwxyz"
def coolEncoder(p):
  cipher_text=""
  for l in p:
    if l in caps:
      v = caps.find(l) + 1
      v = v + 1
      v = padding(v)
      cipher_text += v
    elif l in smalls:
      v = smalls.find(l) + 1
      v = padding(v)
      cipher_text += v
    else:
      v = "0" + str(l)
      cipher_text += v
  return cipher_text

print(coolEncoder("A"))
print(coolEncoder("b"))
print(coolEncoder("2"))
```

The `ord()` function converts a character into its `int` value, and `chr()` does the opposite.

```
a = ord('z')
print(a)

a = chr(45) # a hyphen
print(a)

a=""
for i in range(256):
  a+=chr(i)

print(a)
```

Write an encoder that converts a text string into the character value of its characters, and then write the corresponding decoder.

```
def num_digits(n):
  n = str(n)
  return len(n)

def padding(n):
  if num_digits(n)==2:
    return "0"+str(n)
  elif num_digits(n)==1:
    return "00"+str(n)
  else:
    return str(n)

def better_encoder(p):
  cipher_text=""
  for l in p:
    v=ord(l)
    v=padding(v)
    cipher_text+=v
  return cipher_text

better_encoder("Python")
```
