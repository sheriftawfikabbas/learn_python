## 5.2 Creating and importing modules

We learned, when using functions, that we can re-use our script. However, what if you want to use a set of functions, not just one function? And what if you also want to re-use variables?

Python supports a cool technique for enabling script reusability: *moodules*. A module is quite simple: it's just a python code file. It has to have the python file extension, *.py*, and can contain as many functions and variables as you wish.

Let's create a very simple python script file that has only one variable, `a`, and set `a=1`. One a Windows operating system, you can create a text file  using Notepad and save it on your desktop. Let's call this file: `abc_module.py`.

Once you have done so, start python in the same directory as the one you saved the file. You can `use` your module via the `import` keyword:

```
import abc_module
```

Note that we didn't have to type the file extension - python already assumes it must end with .py.

Then, let's start actually re-using the code! Let's see if we can access `a`:

```
import abc_module

print(abc_module.a)
```

That's how we can access the contents of a module after importing it using the `import` keyword: we use the name of the module, followed by the `.` notation.

There are three dufferent ways to import a module:
1. Using the `import` as above.
2. Using `import module_name as alias`.
3. Using `from module_name import something`. To import everything from `module_name`, we type `from module_name import *`.

In (2), we give the module an alias, or a nick name. For example: 
```
import abc_module as abc

print(abc.a)
```
As for (3), we would use this if we only need some of the things in the module, not the whole module. For example, 

```
from abc_module import a

print(a)
```

Note that when we apply method (3), we don't need to use the module name, or alias, to access the valiable.

So far we've been putting variables into the module. Now let's start putting functions. Copy-paste the following into your Notepad then save it as a abc_module.py:

```
def addition(a,b):
  return a+b

pi = 3.14159265
```

Then write the following in the python interpreter:
```
from abc_module import pi,addition

print(pi)
print(addition(3,4))
```

Here, the value of the variable `pi` and the resturn value of the function will be exectued.

## 5.3 File operations

### Creating a new file

To create a new file in python we use the `open` function:

```
f = open('filename','w')
```

The first argument in this function is the file name, the second is the mode of file access:
- 'w': open the file for a write operation, creates it if it does not exist
- 'a': open the file for an append operation, creates it if it does not exist
- 'r': open the file for a read operation, error if it does not exist
- 'x': creates the file,error if the file exists

### Writing into a file

To write text into a created file, we use the `write()` function:
```
f = open('filename','w')
f.write('hello!')
```

### Closing the file

When you are done with writing and you don't need to access the file any longer in your program, you must close it with the `close()` function:

```
f = open('filename','w')
f.write('hello!')
f.close()
```

### Reading from a file

The `readlines()` function returns a list of each line in the file as a list item.

The code below prints out each line in the file.

```
f = open('filename','w')
l=f.readlines()
for i in l:
   print(i)
f.close()
```
