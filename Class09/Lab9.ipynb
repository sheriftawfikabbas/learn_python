## Exercise 2: (Quiz)

Write a python code that retrieves the JSON object from this address

http://api.open-notify.org/astros

and then prints out the names of the astronauts.

```python
import requests
response = requests.get("http://api.open-notify.org/astros")
data = response.json()

print(data)

l = data['people']
print(l)

for i in l:
    print(i['name'])
```
